# Visualisation of task's 3 result


```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('task3_result.xlsx', header=0, engine='openpyxl')
df.set_index("Continent", inplace=True)
df
```

### Circle diagram of data

```python
df.plot.pie(y="size")
```

### Hystogram of countries' count on continent


```python
df.plot(y=["sum", "mean", "std"], kind="bar")
```

### GDP 2005-2015 by countries

```python
gdp = pd.read_excel('answer1.xlsx', header=0, engine='openpyxl')
# gdp.set_index("Country", inplace=True)
gdp.drop(gdp.iloc[:, 1:11], axis=1, inplace=True)
gdp
```

```python
gdp2 = gdp.melt(id_vars=['Country'], var_name='year', value_name='GDP')
sns.lineplot(x='year', y='GDP', hue='Country', data=gdp2)
```
